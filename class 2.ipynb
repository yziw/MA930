```julia
a1 = 0.5
b1 = 0.5
a2 = 5
b2 = 5
```

```julia
x = range(0, stop=5, length = 1000)
```

```julia
y1 = (b1^a1)*x.^(a1-1.0).*exp.(-b1*x)/gamma(a1);
y2 = (b2^a2)*x.^(a2-1.0).*exp.(-b2*x)/gamma(a2);
```

```julia
# figure(figsize = (20, 8));
# title("hello world");
plot(x, y1, label = "1")
plot!(x, y2, label = "2")
ylims!((0, 2))
```

```julia
n = 5
N = 1000
data1 = rand(Gamma(a1, 1/b1), n, N)
mean1 = zeros(N)
for i = 1:N
    mean1[i] = mean(data1[:, i])
end
```

```julia
mean1hist = fit(Histogram, mean1, nbins = 100)
```

```julia
plot(mean1hist.edges, mean1hist.weights./sum(mean1hist.weights))
```
